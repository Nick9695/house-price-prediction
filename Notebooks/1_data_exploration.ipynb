```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("E:/Git Projects/1 House Price Prediction/Housing.csv")  

# Display basic information
print(df.info())
print(df.describe())

# Check for missing values
print(df.isnull().sum())

# Visualize the distribution of the target variable (price)
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True)
plt.title('Distribution of Sale Prices')
plt.show()

# Explore relationships between numerical features and price
numeric_features = df.select_dtypes(include=['int64', 'float64'])
correlations = numeric_features.corr()['price'].sort_values(ascending=False)
print(correlations)

# Visualize top correlations
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_features[correlations.index[:10]].corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap of Top Correlated Features')
plt.show()

# Explore categorical features
categorical_features = df.select_dtypes(include=['object'])
for column in categorical_features.columns:
    print(f"\n{column}:")
    print(df[column].value_counts())
    print("-" * 50)
```
